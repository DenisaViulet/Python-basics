```python
##ex1
def fizzbuzz(n):
    for i in range(1,n+1):
        if i%10==0:
            print('FuzzBuzz')
        elif i%2==0:
            print('Fuzz')
        elif i%5==0:
            print('Buzz')
        else:
            print(i)
```

```python
n=input('Dati numarul')
n=int(n)
fizzbuzz(n)
```

```python
##ex2-alfabet-scor scrabble
SCORES={"a": 1, "b": 2, "c":4, "d":2,
        "e":3,"f":2,"g":5,"h":5,"i":1,
        "j":3,"k":4,"l":2,"m":1,"n":3,
        "o":5,"p":1,"q":6,"r":3,"s":2,
        "t":1,"u":5,"v":3,"w":3,"x":4,
        "y":5,"z":6}  ##alfabetul
def scrabbleScore(word):
    total=0
    for letter in word:
        letter=letter.lower()
        total+=SCORES[letter]
    return total
print(scrabbleScore("LaboRator"))

# sau
# def scrabbleScore(word)
#     return sum(SCORES[letter.lower()] for letter in word))
```

```python
##ex3-Anagrame
def isAnagram(word1,word2):
    return sorted(word1)==sorted(word2)
word1=input('Enter the first string: ')
word2=input("Enter the second string: ")
if isAnagram(word1,word2):
    print('Are anagram')
else:
    print('Are not anagram ')
```

```python
##ex4-conjectura Collatz
def Collatz(n):
    count=0
    while n!=1:
        if n%2==0:
            n = n/2
            count+=1
        else:
            n = 3*n+1
            count+=1
    return count
    
for i in range(1,1001):
    print(Collatz(i))
```

```python
##ex5- permutari circulare
def isCircular(arr1,arr2):
    if len(arr1) != len(arr2):
        return False
    str1=' '.join(map(str,arr1))
    str2=' '.join(map(str,arr2))
    if len(str1) != len(str2):
        return False
    return str1 in str2+ ' ' +str2
arr1=[1,2,3,4]
arr2=[3,4,1,2]
print (isCircular(arr1,arr2))
```
